She Walks in Beauty by Lord Byron

```python
import os

os.chdir('english')

print(os.listdir()) 

with open('she_walks_in_beauty.txt', 'r', encoding='utf-8') as file:
    text = file.read()

print(text[:300]) 
```

### Токенизация и очистка текста
В этой ячейке мы выполняем токенизацию текста и удаляем стоп-слова, чтобы подготовить данные для дальнейшего анализа.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# nltk.download('punkt')
# nltk.download('stopwords')

tokens_english = word_tokenize(text.lower()) 

stop_words_english = set(stopwords.words('english'))
filtered_tokens_english = [word for word in tokens_english if word.isalpha() and word not in stop_words_english]

print(filtered_tokens_english[:20])
```

создание WordCloud и графика для английского текста

```python
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

counter_eng = Counter(['walks', 'beauty', 'like', 'night', 'cloudless', 'climes', 'starry', 'skies', 'best', 'dark', 'bright', 'meet', 'aspect', 'eyes', 'thus', 'mellowed', 'tender', 'light', 'heaven', 'gaudy'])  
most_common_eng = counter_eng.most_common(10)

words_eng, counts_eng = zip(*most_common_eng)

plt.figure(figsize=(10, 5))
plt.bar(words_eng, counts_eng, color='skyblue')
plt.title('Top 10 Most Frequent English Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


wordcloud_eng = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(counter_eng)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud_eng, interpolation='bilinear')
plt.axis('off')
plt.title("English Word Cloud")
plt.tight_layout()
plt.show()

wordcloud_eng.to_file('english_wordcloud.png')
```

Произведение "Красавица" от А.С.Пушкин

```python
os.chdir('..') 
os.chdir('russian')

print(os.listdir())

with open('krasavitse.txt', 'r', encoding='utf-8') as file:
    text_russian = file.read()

print(text_russian[:300])
```

### Токенизация и очистка текста
В этой ячейке мы выполняем токенизацию текста и удаляем стоп-слова, чтобы подготовить данные для дальнейшего анализа.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# nltk.download('stopwords')
# nltk.download('punkt')

tokens = word_tokenize(text_russian.lower())  # Преобразуем в нижний регистр для единого стандарта

stop_words = set(stopwords.words('russian'))
filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

print(filtered_tokens[:20])

```

создание WordCloud и графика для русского текста

```python
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

counter_rus = Counter(['красавице', 'которая', 'нюхала', 'табак', 'возможно', 'ль', 'вместо', 'роз', 'амуром', 'насажденных', 'тюльпанов', 'гордо', 'наклоненных', 'душистых', 'ландышей', 'ясминов', 'лилей', 'которых', 'любила', 'прежде'])
most_common_rus = counter_rus.most_common(10)

words_rus, counts_rus = zip(*most_common_rus)

plt.figure(figsize=(10, 5))
plt.bar(words_rus, counts_rus, color='skyblue')
plt.title('Top 10 Most Frequent Russian Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

wordcloud_rus = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(counter_rus)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud_rus, interpolation='bilinear')
plt.axis('off')
plt.title("Russian Word Cloud")
plt.tight_layout()
plt.show()

wordcloud_rus.to_file('russian_wordcloud.png')
```
